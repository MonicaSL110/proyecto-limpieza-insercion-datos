```python
import re
import os
import xml.etree.ElementTree as ET

tree = ET.parse('data_xml.xml')
root = tree.getroot()


xml_list = []
for child in root:
    xml_dict = {}
    for subchild in child:
        xml_dict[subchild.tag] = subchild.text
    xml_list.append(xml_dict) # Es una lista que contiene diccionarios
```

```python
xml_list
```

```python
#eliminamos la columna level_0

for diccionario in xml_list:
    
    del (diccionario['level_0'])   
    
xml_list
        
```

```python
#cambiamos la variable género 0=man, 1=woman, 2=non binary, 3=prefer not to say, 4=prefer to self describe

for diccionario in xml_list:
    if diccionario.get('gender') == '0':
        print('man')
    elif diccionario.get('gender') == '1':
        print('woman')
            
```

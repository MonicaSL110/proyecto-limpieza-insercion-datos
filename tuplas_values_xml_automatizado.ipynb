```python
import re
import os
import xml.etree.ElementTree as ET

tree = ET.parse('data_xml.xml')
root = tree.getroot()


xml_list = []
for child in root:
    xml_dict = {}
    for subchild in child:
        xml_dict[subchild.tag] = subchild.text
    xml_list.append(xml_dict) # Es una lista que contiene diccionarios
```

```python
xml_list
```

```python
def conversion_datos_xml(datos):
    """ 
    Esta función convierte una lista de diccionarios en una lista de diccionarios con los datos limpios.

    Args: 
        - datos (list): la lista de diccionarios con todos los datos que se limpiearan e insertaran.

    Returns:
        Devuelve una lista de diccionarios limpia.
    """ 
    data_xml=[]
    for diccionario in datos:
        if diccionario.get('gender') == '0':
            diccionario.update({'gender':'Man'})
        elif diccionario.get('gender') == '1':
            diccionario.update({'gender':'Woman'})        
        elif diccionario.get('gender') == '2':
            diccionario.update({'gender':'Nonbinary'})
        elif diccionario.get('gender') == '3':
            diccionario.update({'gender':'Prefer not to say'})        
        elif diccionario.get('gender') == '4':
            diccionario.update({'gender':'Prefer to self-describe'})
        del (diccionario['level_0'])
        data_xml.append(diccionario)
        
    return data_xml
```

```python
datos_convertidos_xml = conversion_datos_xml(xml_list)
```

```python
def values_tupla(datos_convertidos):
    """ 
    Esta función convierte una lista de diccionarios en una lista de tuplas con los datos limpios.

    Args: 
        - datos (list): la lista de diccionarios con todos los datos de los cuales se extraerán las values.

    Returns:
        Devuelve una lista de tuplas con solo los valores de las values.
    """ 
    lista = []
    for diccionario in datos_convertidos:
        tupla = ()
        indice = 0
        for i in diccionario:
            tupla = tupla + (diccionario[i],)
            indice += 1
            if indice == len(diccionario):
                lista.append(tupla)
    
    lista_final = []

    for i in lista:
        lista_tupla = list(i)
        lista_tupla.append(lista_tupla[0])
        tupla_final = tuple(lista_tupla)
        lista_final.append(tupla_final)
        
    return lista_final
```

```python
values_tupla = values_tupla(datos_convertidos_xml)
```
